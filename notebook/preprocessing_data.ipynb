Exploratory Data Analysis (EDA)

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Visualize the Fraud_Data.csv file

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='class', data = df_fraud)
plt.title('Distribution of Fraudulent Transactions')
plt.xlabel('Class (0: Non-Fraud, 1: Fraud)')
plt.ylabel('Count')
plt.show()
```

```python
df_fraud.columns
```

User Demographics - Age Distribution
Visualizing the age distribution of users can provide insights into demographics related to fraud.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_fraud['age'], bins=30, kde=True)
plt.title('Age Distribution of Users')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

Purchase Value Analysis
Understanding the distribution of purchase values helps identify patterns that may indicate fraud.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='class', y='purchase_value', data=df_fraud)
plt.title('Purchase Value Distribution by Class')
plt.xlabel('Class (0: Non-Fraud, 1: Fraud)')
plt.ylabel('Purchase Value ($)')
plt.show()
```
