```python
import ipaddress

def is_valid_ip(ip):
    try:
        # Validate the IP address
        ipaddress.ip_address(ip)
        return True
    except ValueError:
        # Return False if the IP address is invalid
        return False

# Filter the dataset to keep only valid IP addresses
fraud_data = df_fraud[df_fraud['ip_address'].apply(is_valid_ip)]
def ip_to_int(ip):
    # Convert IP address string to an integer
    return int(ipaddress.ip_address(ip))

# Apply the conversion function to the 'ip_address' column
fraud_data['ip_int'] = fraud_data['ip_address'].apply(ip_to_int)
```

```python
import ipaddress

def int_to_ip(ip_int):
    # Convert integer IP to dotted-decimal format
    return str(ipaddress.ip_address(int(ip_int)))

# Apply the conversion to the 'ip_address' column
df_fraud['ip_address_str'] = df_fraud['ip_address'].apply(int_to_ip)

# Now filter using the is_valid_ip function
fraud_data = df_fraud[df_fraud['ip_address_str'].apply(is_valid_ip)]

# Convert the valid IPs to integer format for merging later
fraud_data['ip_int'] = fraud_data['ip_address_str'].apply(ip_to_int)

# Display the number of rows after filtering
print(f"Number of rows after filtering valid IP addresses: {len(fraud_data)}")
```

```python
# Sort the DataFrames by the relevant columns before merging
fraud_data = fraud_data.sort_values('ip_int')

```

```python
import ipaddress

def int_to_ip(ip_int):
    # Convert integer to IP address string
    return str(ipaddress.ip_address(int(ip_int)))

# Convert lower and upper IP bounds from integer format to IPv4 string format
df_ipaddres['lower_bound_ip_address'] = df_ipaddres['lower_bound_ip_address'].apply(int_to_ip)
df_ipaddres['upper_bound_ip_address'] = df_ipaddres['upper_bound_ip_address'].apply(int_to_ip)
def ip_to_int(ip):
    # Convert IP address string to an integer
    return int(ipaddress.ip_address(ip))

# Convert lower and upper IP bounds to integer format
df_ipaddres['lower_bound_ip_int'] = df_ipaddres['lower_bound_ip_address'].apply(ip_to_int)
df_ipaddres['upper_bound_ip_int'] = df_ipaddres['upper_bound_ip_address'].apply(ip_to_int)

```

```python
ip_to_country = df_ipaddres.sort_values('lower_bound_ip_int')
```

```python
# Step 1: Sort the DataFrames by the relevant columns before merging
fraud_data = fraud_data.sort_values('ip_int')
ip_to_country = df_ipaddres.sort_values('lower_bound_ip_int')

# Step 2: Merge using pd.merge_asof to match the IP address within the IP ranges
# This will match each row in fraud_data with the closest lower_bound_ip_int in ip_to_country
merged_data = pd.merge_asof(
    fraud_data,
    ip_to_country,
    left_on='ip_int',
    right_on='lower_bound_ip_int',
    direction='backward',
    suffixes=('', '_country')
)

# Step 3: Filter the merged DataFrame to ensure that 'ip_int' falls within the IP range
# We check that 'ip_int' is between 'lower_bound_ip_int' and 'upper_bound_ip_int'
merged_data = merged_data[
    (merged_data['ip_int'] >= merged_data['lower_bound_ip_int']) &
    (merged_data['ip_int'] <= merged_data['upper_bound_ip_int'])
]
```
